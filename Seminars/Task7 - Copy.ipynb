```python
X, y = datasets.make_classification(
    n_samples=20,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2, 
    random_state=8
)
y = np.where(y == 0, -1, 1)

plt.figure(figsize=(7, 6))
class_0 = X[y == -1]
class_1 = X[y == 1]
plt.scatter(class_0[:, 0], class_0[:, 1], label='-1 класс')
plt.scatter(class_1[:, 0], class_1[:, 1], label='1 класс')
plt.legend()
plt.xlabel('X1')
plt.ylabel('X2')

w1, w2 = 1, 0.2
w = np.array([w1, w2])

def calculate_margin(x, w):
    return abs(np.dot(w, x)) / np.linalg.norm(w)

for  (x_coord, y_coord) in X:
    margin = calculate_margin((x_coord, y_coord), w)
    plt.text(x_coord, y_coord, f"{margin:.2f}", fontsize=8, ha='right', color='black')


x1_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 1000)
x2_vals = -(w1 * x1_vals) / w2
plt.plot(x1_vals, x2_vals, color="green", label='Decision Boundary')

#plt.scatter(w1, w2, color="red", s=100, label='Weight Vector (w)')
plt.xlim(-2, 3)
plt.ylim(-2, 3)
#plt.legend()

plt.grid(True)
plt.show()
```

```python
pred_df = pd.DataFrame({
    'y Prediction': np.sign(X @np.array([w1,w2])),
    'y Margin':np.where(np.sign(y*(X @np.array([w1,w2]))) == -1, False, np.where(np.sign(y*(X @np.array([w1,w2]))) == 1, True, None)),
    'y True': y,
    
})

pred_df
```
